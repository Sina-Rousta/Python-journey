```python
def print_board(inputs):
    print('        |         |        ')
    print(f'   {inputs[6]}    |    {inputs[7]}    |   {inputs[8]}   ')
    print('        |         |        ')
    print('---------------------------')
    print('        |         |        ')
    print(f'   {inputs[3]}    |    {inputs[4]}    |   {inputs[5]}   ')
    print('        |         |        ')
    print('---------------------------')
    print('        |         |        ')
    print(f'   {inputs[0]}    |    {inputs[1]}    |   {inputs[2]}   ')
    print('        |         |        ')
    
```

```python
def welcome():
    answer1 = input("Hello, welcome to Tic Tac Toe! Enter 'Play' to begin or 'Exit' to quit: ")
    while answer1.lower() not in ['play','exit']:
        clear_output()
        answer1 = input("Hello, welcome to Tic Tac Toe! Enter 'Play' to begin or 'Exit' to quit: ")
        
    if answer1.lower() == 'play':
        return True
    else:
        return False
```

```python
def play_again():
    answer2 = input("Enter 'Yes' to play again or 'No' to exit: ")
    while answer2.lower() not in ['yes','no']:
        clear_output()
        answer2 = input("Enter 'Yes' to play again or 'No' to exit: ")
        
    if answer2.lower() == 'yes':
        return True
    else:
        return False
```

```python
from IPython.display import clear_output

def choose_marker():  
    marker1 = input('Dear player 1, do you want to be X or O? ')
    
    while marker1 not in ['X','O']:
        clear_output()
        marker1 = input('Invalid Choice! Please enter X or O ')
    clear_output()
    
    if marker1 == 'X':
        marker2 = 'O'
    else:
        marker2 = 'X'
    
    print(f"So player 1 is {marker1} and player 2 is {marker2}")
    
    return [marker1,marker2]
```

```python
def players_choice():
    acceptable_range = range(1,10)
    
    result = input('Enter your number (1-9): ')
    
    while result.isdigit() == False or int(result) not in acceptable_range:
        result = input('Please enter a valid value: ')
    
    
    return int(result)
```
